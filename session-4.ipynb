```python
import nltk
from nltk.corpus import names
```

```python
names.fileids()
```

```python
len(names.words())
```

```python
labeled_names = ([(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
print(labeled_names)
```

```python
def gender_features(word):
    return {'last letter' : word[-1]}
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
print(featuresets)
```

```python
train_set,test_set = featuresets[5000:],featuresets[:2000]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('David'))
```

checking the accuracy for the last 2000 data in test_set 

```python
print(nltk.classify.accuracy(classifier,test_set))
```

CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(binary=True)
corpus = ["Tessaract is good optical  character recognition software", "optical character recognition is significant"]
vect.fit(corpus)

vocab = vect.vocabulary_ #vocabulary_ built-in:-
for key in sorted(vocab.keys()):
    print("{}:{}",format(key,vocab[key]))
```
